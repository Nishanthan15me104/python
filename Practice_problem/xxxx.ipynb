```python
def calculate_accuracy(predicted_labels, actual_labels):
  """Calculates the accuracy of a binary classification model.

  Args:
    predicted_labels: A list of predicted labels.
    actual_labels: A list of actual labels.

  Returns:
    The accuracy of the model.
  """

  correct_predictions = 0
  total_predictions = len(predicted_labels)

  for predicted_label, actual_label in zip(predicted_labels, actual_labels):
    if predicted_label == actual_label:
      correct_predictions += 1

  accuracy = correct_predictions / total_predictions
  return accuracy

# Sample data
predicted_labels = [1, 0, 1, 1, 0]
actual_labels = [1, 1, 0, 1, 0]

# Calculate accuracy
accuracy = calculate_accuracy(predicted_labels, actual_labels)
print("Accuracy:", accuracy)
```

```python
import numpy as np
import pandas as pd

def categorize_sales(sales):
    """Categorizes sales data based on percentiles.

    Args:
        sales (float): The sales value to categorize.

    Returns:
        str: The category of the sales value ('low seller', 'top seller', or 'average seller').
    """

    sales_20percentile = np.percentile(sales, 20)
    sales_80percentile = np.percentile(sales, 80)

    if sales < sales_20percentile:
        return 'low seller'
    elif sales > sales_80percentile:
        return 'top seller'
    else:
        return 'average seller'

# Sample DataFrame
data = {'sales': [100, 200, 50, 300, 150, 250, 75, 350, 200, 175]}
df_sales = pd.DataFrame(data)

# Apply the categorization function to the 'sales' column
df_sales['categories_sales'] = df_sales['sales'].apply(categorize_sales)

print(df_sales)
```

```python
import pandas as pd
import numpy as np



# Load the employee salaries dataset
df = pd.read_csv('employee_salaries.csv')

# Define the salary threshold
salary_threshold = 50000

# Initialize total salary
total_salary = 0

# Set the index to start iterating
i = 0

# While loop to iterate through the dataframe
while i < len(df):
    salary = df.loc[i, 'Salary']
    if salary > salary_threshold:
        total_salary += salary
    i += 1

# Print the result
print(f"Total salary paid to employees earning above {salary_threshold} is: ${total_salary}")
```

```python
import pandas as pd

# Load the student grades dataset
df = pd.read_csv('student_grades.csv')

# Define passing marks
passing_marks = 60

# Initialize counter
passed_students = 0

# Set the index to start iterating
i = 0

# While loop to iterate through the dataframe
while i < len(df):
    grade = df.loc[i, 'Grade']
    if grade >= passing_marks:
        passed_students += 1
    i += 1

# Print the result
print(f"Number of students who passed: {passed_students}")
```

```python
import pandas as pd

# Load the stock prices dataset
df = pd.read_csv('stock_prices.csv')

# Initialize the count of days with closing price higher than opening price
count_higher_close = 0

# Set the index to start iterating
i = 0

# While loop to iterate through the dataframe
while i < (len(df) -1):
    open_price = df.loc[i, 'Close']
    close_price = df.loc[i+1, 'Close']
    if close_price > open_price:
        count_higher_close += 1
    i += 1

# Print the result
print(f"Number of days where the stock closed higher than it opened: {count_higher_close}")
```
